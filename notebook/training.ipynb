```python
# Import Neccessary libraries
import numpy as np
import pandas as pd
import mlflow

# Import Visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Import for Model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, classification_report,f1_score, roc_curve, roc_auc_score, precision_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

#Import Sampler libraries
from imblearn.combine import SMOTEENN
from imblearn.over_sampling import RandomOverSampler
```

```python
mlflow.set_tracking_uri('http://mlflow-web:5000')
```

```python
mlflow.set_experiment("experiment_diabetes")
```

```python
mlflow.autolog()
```

```python
df = pd.read_csv('data/data-cleaned.csv')
df
```

```python
x = df.drop(columns='Outcome')
y = df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_test
```

## Create Model

```python
with mlflow.start_run():
    model = RandomForestClassifier(random_state=42, n_estimators=200, max_depth=20, min_samples_split=2, min_samples_leaf=1, max_features='sqrt',criterion='gini', class_weight='balanced_subsample', bootstrap=False)
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    mlflow.log_metric("accuracy", accuracy)
```
